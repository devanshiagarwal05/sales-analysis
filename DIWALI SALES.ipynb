```python
a=sns.countplot(x="Gender",data=df)
for bars in a.containers:
    a.bar_label(bars)
```

```python
df.groupby(["Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
```

```python
sns.countplot(data=df,x="Age Group",hue="Gender")
```

```python
sns.countplot(data=df,x="Age Group")
```

```python
a=sns.countplot(data=df,x="Age Group",hue="Gender")
for bars in a.containers:
    a.bar_label(bars)
```

```python
v=df.groupby(["Age Group"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.barplot(x="Age Group",y="Amount",data=v)
```

```python
# buyers of age 26-35 has done most shopping
```

```python
# want to know order numbers from specific states
```

```python
f=df.groupby(["State"],as_index=False)["Orders"].sum().sort_values(by="Orders",ascending=False).head(10)#head is basically used for specifying top(10),therfore number can be changed
sns.set(rc={"figure.figsize":(25,10)})
sns.barplot(x="State",y="Orders",data=f)
```

```python
# sabse zyada order uttar pradesh se hue hai
```

```python
f=df.groupby(["State"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False).head(10)#head is basically used for specifying top(10),therfore number can be changed
sns.set(rc={"figure.figsize":(25,10)})
sns.barplot(x="State",y="Amount",data=f)
```

```python
#this we have done basically that we can se (kerela) state is dismissed as it has done more order but not generated that much amount
```

```python
sns.set(rc={"figure.figsize":(3,5)})
a=sns.countplot(data=df,x="Marital_Status")
for bars in a.containers:
    a.bar_label(bars)
```

```python
f=df.groupby(["Marital_Status","Gender"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(3,4)})
sns.barplot(x="Marital_Status",y="Amount",data=f,hue="Gender")
```

```python
# married women has spent most amount
```

```python
#now I wanna see the relation between amount and occupation means which occupation spends more on shopping
```

```python
f=df.groupby(["Occupation"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)
sns.set(rc={"figure.figsize":(25,10)})
sns.barplot(x="Occupation",y="Amount",data=f)
```

```python
# thus as we can see the most purchasing power is from it sector
```
